```python
# Importing all libraries required in this notebook
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

The pandas library is used for data manipulation and analysis. Numpy is used for numerical computations, and seaborn and matplotlib.pyplot are used for data visualization. The scikit-learn library is used for machine learning, specifically for splitting the data into training and testing sets and for fitting a linear regression model.

```python
# Reading data github

df = pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
df
```

```python
sns.scatterplot(data=df, x='Hours', y='Scores').set(title='Percentage scored with respect to study hours')
```

This Scatter plot is used to determine whether or not two variables have a relationship or correlation or not. It is plotted between Hours student studied and Percentage student scored. These two variables are positively correlated with each other

```python
a=df.corr()
sns.heatmap(a,annot=True)
```

Heatmap checks the correlation between the variables This method is useful to determine the strength and direction of the linear relationship between the different columns in a dataframe. It can help identify which features are strongly correlated with the target variable, and also identify any multicollinearity between the features.

```python
sns.pairplot(df)
```

This plot is useful for identifying patterns and relationships between variables. It also helped to identify any outliers or data points that do not fit the general pattern.In this dataset there are not outliers

```python
sns.distplot(df.Scores)
```

```python
sns.distplot(df.Hours)
```

This plot is useful for visualizing the distribution of a variable and identifying its central tendency and spread.

```python
x=df["Hours"]
y=df["Scores"]
```

Here x is independent variable and y is dependent variable

```python
x_ = df.iloc[:, :1].values  
y_ = df.iloc[:, 1].values  
```

This creates the feature matrix for machine learning models. In this case, x_ contains the values of a single feature or predictor variable.

```python
x_train,x_test,y_train,y_test=train_test_split(x_,y_,test_size=0.25,random_state=0)
```

spliting the dataset into training and testing sets for machine learning models.Here we are splitting two arrays x_ and y_ into training and testing sets, with 75% of the data allocated to training and 25% to testing.

```python
x_train
```

```python
reg=LinearRegression()
reg.fit(x_train,y_train)
print("Training completed.")
```

It fits the linear regression model to the training data. The fit method gives the coefficients of the linear regression equation using the training data. x_train and y_train are the training sets of the predictor and response variables.

```python
print('slope',+reg.coef_)
print('intercept',+reg.intercept_)
```

```python
# Plotting the regression line
line = reg.coef_*x+reg.intercept_

# Plotting for the test data
plt.scatter(x, y)
plt.plot(x, line);
plt.show()
```

'line' calculates the equation of the regression line(best fit line) based on the coefficients estimated by the LinearRegression model.This line can be used to visualize the relationship between the dependent and independent variable in the data, and to make predictions of the dependent(target) variable for new values of the independent variables.

```python
y_pred=reg.predict(x_test)
df = pd.DataFrame({'Actual values': y_test, 'Predicted values': y_pred})  
df 
```

```python
# Testing with new data
t=9.2
t_pred = reg.predict([[t]])
print('Hours Studied',+t)
print('Predicted percentage',+t_pred)
```

```python
from sklearn import metrics  
print('Mean Absolute Error:', 
      metrics.mean_absolute_error(y_test, y_pred)) 
```

mean absolute error (MAE) between the predicted values and the actual values of the response variable in the testing set.
